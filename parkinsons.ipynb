# Parkinsons Analysis


## Import Data

```python
import pandas as pd
%matplotlib inline
df = pd.read_csv('parkinsons.data')
df.head()
```

## Clean up data (if necessary)

```python
df.isnull().values.any()

all_features = df.drop(['status'],axis=1) 
```

Do some plotting to visuallise some information


```python
import matplotlib.pyplot as plt


def scatter_plot(x_name, y_name, title):
    x_data = df[x_name]
    y_data = df[y_name]
   
    scatter = plt.scatter(x_data, y_data, c = df['status'])
    plt.title(title)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.legend(*scatter.legend_elements())
    plt.show()

scatter_plot('MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', "classification of parkinsons per MDVP:Fhi(Hz) nad MDVP:Flo(Hz)")
```

```python
scatter_plot('MDVP:Fo(Hz)','status', "Avg VFF vs. parkinsons")
```

We can seeï¼š
- People who have parkinsons tend to have lower maximum VFF and higher maximum average VFF

## Variations of VFF ~ Parkinsons

```python
plt.ylim(0, 0.00027)
plot = scatter_plot("MDVP:Jitter(%)","MDVP:Jitter(Abs)", "classcification of parkinsons in Jitter in VFF")
```

```python
scatter_plot("MDVP:RAP","status", "classcification of parkinsons by using MVDP:RAP")
```

We can see:
-  people who have parkinsons tend to have lower variations in frequency

## Parkinsons ï½ž Variation in amplitude


```python
scatter_plot("MDVP:Shimmer","MDVP:Shimmer(dB)", "classcification of parkinsons by MVDP:Shimmer and MVDP:shimmer(dB)")
```

```python
scatter_plot("Shimmer:DDA", "status","")
```

We can see:
-  people who have parkinsons tend to have lower variations in amplitude

```python
scatter_plot("NHR","HNR", "classification of parkinsons by measuring NHR HNR")
```

We can see:
- People who has parkinsons tend to have higher HNR but lower NHR
